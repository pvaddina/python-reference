# Reading CSVs

```python
appst = pd.read_csv("data/appstore_games.csv")
appst.head()
```

## Get minimum and maximum values of a column

```python
print("Minimum price={}, and Maximum price={}".format(appst.Price.min(), appst.Price.max()))
```

## Filtering data based on a condition

```python
# Get all the apps which have a specific "Age Rating" ..
kids_apps = appst[appst["Age Rating"] == "4+"]
kids_apps.head()
```

```python
# Get all the apps that are free of cost
free_apps = appst[appst.Price <= 0]
free_apps
```

```python
# Get all the apps that are badly reviewed and yet are priced more than 2
one_rated_apps = appst[appst["Average User Rating"] <= 1]
crappy_apps = one_rated_apps[one_rated_apps["Price"] >= 2]
crappy_apps
```

```python
# Plot a graph with "User Rating Count" on the x-axis and "Price" on the y-axis
plt.plot(crappy_apps["User Rating Count"], crappy_apps["Price"], "o")
plt.plot(crappy_apps["User Rating Count"], crappy_apps["Price"])
plt.xlabel("User Rating Count")
plt.ylabel("Price")
plt.show
```
