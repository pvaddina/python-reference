```python
import pandas as pd
import numpy as np
```

```python
dfile = "../data/appstore_games.csv"
```

```python
apps = pd.read_csv(dfile, encoding="utf8")
apps.head()
```

```python
# 
# Check the number of entries where the "Current Version Release Date" is unusually older than 
# the "Original Release Date". 
# 
apps['Original Release Date'] = pd.to_datetime(apps['Original Release Date'])
apps['Current Version Release Date'] = pd.to_datetime(apps['Current Version Release Date'])

# Add a new column which gives the number of days between the 'Original release date' 
# and the 'Current version release date'
apps['DaysOfExistence'] = apps['Current Version Release Date'] - apps['Original Release Date']
apps[apps.DaysOfExistence < pd.Timedelta(0,'D')]
```

```python
# Get the total number of items where the current version date is unusually older than the original release date
apps.DaysOfExistence[apps.DaysOfExistence < pd.Timedelta(0,'D')].count()
```

```python
# Lets find out how many unique 'Primary Genre' are there in the DS
apps['Primary Genre'].unique()
```

```python
# Get the number of apps that come under the 'Stickers' Genre
apps[apps['Primary Genre']=='Stickers'].count()
```

```python
apps[apps['Primary Genre']=='Stickers']['Primary Genre'].count()
```

```python
# Get all the statistics of the apps 'ID' column when the data is grouped according to the 'Primary Genre' 
# (ex: count, mean min, max, etc...). Note that in this observation we have a min, max, mean, etc for the 
# 'ID' column which does not make any sense. But when observed for other columns might make sense. 
#
# Unsurprisingly the number of Apps that are under the 'Games' genre are too many compared to the other genres
# as the dataset is about App store games
# 
apps.groupby('Primary Genre')['ID'].describe()
```

The best App in the store based on the average user rating count and the number of users who have rated the App

```python
apps.groupby(['Primary Genre', 'Average User Rating'])['User Rating Count'].describe()
```

```python
# Find the 'maximum' number of 'User Rating Count' among all the App-genres with respect to 
# the different 'Average User Rating'

maxRatings = apps.pivot_table('User Rating Count', index='Primary Genre', columns='Average User Rating', aggfunc='max')
maxRatings
```

We can now infer that for the Genre 'Games' there is atleast one App, which managed to get an average rating of 5.0 despite  at least 374772.0 people rating the App, indicating the APP is very popular among several Users

```python
# What is the most popular game in the App store ?
# An app can be termed as most popular if it has the best rating despite many people commenting on it

maxUserRatings = maxRatings[5.0].max()
```
