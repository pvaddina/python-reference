# Applying custom functions

```python
def range_enumeration(val):
    if val >= 0 and val <3:
        return "low"
    elif val >=3 and val <6:
        return "mid"
    elif val >=6 and val <=9:
        return 'high'
```

```python
pf['range'] = pf['four'].apply(range_enumeration)
pf
```

```python
map_dict = {'mid':500, 'low':0, 'high':1000}
pf['new_range'] = pf['range'].map(map_dict)
pf
```

```python
pf['range'] = pf['range'].map(map_dict) # Note that only a single call will result in an ideal value 
pf
```
