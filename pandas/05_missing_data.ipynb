```python
import numpy as np
import pandas as pd
```

```python
axis1 = ['A','B','C','D','E']
axis0 = [10, 20, 30, 40, 15, 6]
np.random.seed(437)
data  = np.random.randint(1,100,(6,5))
data
```

```python
t = pd.DataFrame(data,axis0, axis1)
t
```

```python
t.iloc[0,1]
```

```python
t.loc[20,['C','D']] = np.nan
t.loc[[20,40], ['A','E']] = np.nan
t
```

```python
t.sort_index(ascending=False)
```

# Dropping missing data

```python
t.dropna()
```

```python
t.dropna(axis=0)
```

```python
t.dropna(axis=1)
```

```python
t.dropna(axis=1, thresh=3) # Atleast 3 non-NaN values
```

# Filling the missing data

```python
t.fillna(0) # Fill all the NaN values to 0
```

```python
t.fillna(t.mean())
```

```python
t.fillna(t.loc[20].mean())
```
