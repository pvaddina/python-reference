```python
np.random.seed(4)
p = pd.DataFrame({"Key": ['a', 'b', 'c', 'c', 'd', 'a', 'b', 'a', 'd'], "Value": np.random.randint(1,10,9)})
q = pd.DataFrame({"Key": ['c', 'a', 'd', 'e', 'a', 'b', 'd', 'e', 'c'], "Value": [3, 5, None, 8, 3, None, 4, 6, 8], "Name": ["One", "Two", "One", "Three", "Four", "Four", "Three", "Two", "One"], "Quotient": np.random.randint(4324,98834827,9)})
print("p=\n", p, "\n\nq=\n", q)
```

```python
# Get a groupby object from the DataFrame
gbP = p.groupby('Key')
gbQ = q.groupby('Key')

print('P, sum()=\n', gbP.sum(), '\nprod()=\n', gbP.prod(), '\nmin()\n', gbP.min(), '\nmax()\n', gbP.max(), '\nmean()\n', gbP.mean(), '\nmedian()\n', gbP.median())
print('q, sum()=\n', gbQ.sum(), '\nq, prod()=\n', gbQ.prod())

print('\n')

# Iterate through all the groups and applying the aggregation functions
for (k,g) in q.groupby('Key'):
    print("Key={}, mean=\n{}\nsum=\n{}\n".format(k, g.mean(), g.sum()))

# Notice how the callable functions are passed as a list
print(q.groupby('Key').aggregate([np.mean, "max", np.max, "min"]))

# Notice how the callable functions are passed as a list
q.groupby('Key').aggregate({"Value": "sum", "Quotient": "min"})
```

```python
# Filtering techniques
def filter_fn(x): 
    return x['Value'].sum() != 0

# This variation with a lambda does not work
# q.groupby(Key').filter(lambda x: return x["Value"].sum() != 0)

# This works
q.groupby('Key').filter(filter_fn)
```

```python
q.groupby('Key').transform(lambda x: x + x.mean())
```

```python
# The dataset contains the year this discovery was made. Lets first convert them 
# to represent decades rather than concrete years. 
# Note here the column 'year' is being modified and given a name 'decade'
decade = (10*(planets['year']//10))
# Cast the decades of numbers to strings and add an 's' to all the series elements
decade = decade.astype(str) + 's'
# Check the type of the object
type(decade)
```

```python
# Give a name to the object
decade.name = 'decade'
# Now group the data based on 'method' and decade series
planets.groupby(['method', decade])['number'].sum().unstack().fillna(0)
```
