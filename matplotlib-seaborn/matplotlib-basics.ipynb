```python
cities = ['Munich', 'Berlin', 'Stuttgart', 'Hamburg', 'Dresden', 'Cologne']
house_types = ["Studio", 'Flat', 'Row house', 'Single house', 'Villa']
np.random.seed(437)
values = np.random.randint(600,2000,(5,6))
p = pd.DataFrame(values, columns=cities)
p['HouseTypes'] = house_types
p
```

```python
plt.plot(p['HouseTypes'], p['Berlin'])
```

```python
plt.scatter(p['HouseTypes'], p['Munich'])
```

```python
plt.plot(p['HouseTypes'], p['Dresden'], color='green', marker= '*')
plt.xlabel('House type')
plt.ylabel('Price in euros')
plt.ylim(500,2000)
```
