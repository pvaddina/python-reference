```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pf = pd.read_csv("../data/heart.csv")
pf.head()
```

```python
sns.distplot(pf['age'])
```

```python
sns.distplot(pf['age'], bins=3)
```

```python
sns.distplot(pf['age'], bins=30, kde=False)
```

```python
plt.figure(figsize=(15,5))
sns.distplot(pf['age'], bins=20, color='green')
```

```python
sns.countplot(pf['sex']) # Essentially plotting number of 'males' vs 'female'
```

```python
plt.figure(figsize=(30,5))
sns.countplot(pf['age'])
```

```python
plt.figure(figsize=(30,5))
sns.countplot(x='age', data=pf)
```

```python
plt.figure(figsize=(30,5))
sns.countplot(x='age', hue='sex', data=pf)
```

```python
sns.boxplot(x='sex', y='age', data=pf)
```

```python
sns.boxplot(x='target',y='age',data=pf,hue='sex')
```

```python
sns.scatterplot(x='chol', y='thalach', data=pf, hue='sex', palette='Dark2', size='age')
```
